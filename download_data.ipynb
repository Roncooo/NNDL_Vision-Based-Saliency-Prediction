# Download Data from Google Drive
This notebook downloads the required data zip file from Google Drive and unzips it.

```python
import gdown
import zipfile
import glob
import os
```

```python
# The ID of the main Google Drive folder you provided
folder_id = '1ifPMdl0ikg8Wre7xRR4rSEA0vGU08Y_k'
output_folder = 'drive_folder/'
```

```python
# Download the entire folder automatically
print("Downloading folder from Google Drive...")
gdown.download_folder(id=folder_id, output=output_folder, quiet=False)
```

```python
# Unzip all downloaded zip files
print("Extracting files...")

# Find all zip files downloaded in the folder (and subfolders)
zip_files = glob.glob(os.path.join(output_folder, '**', '*.zip'), recursive=True)

for zip_file in zip_files:
    print(f"Extracting {zip_file}...")
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(output_folder)
    
print("Extraction complete!")
```
